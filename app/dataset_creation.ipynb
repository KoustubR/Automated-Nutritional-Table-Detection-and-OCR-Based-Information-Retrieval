```python
df = pd.read_csv('../csv/ireland.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.shape
```

```python
sampled_df = df.sample(n=1000, random_state=23011475)
```

```python
sampled_df.to_csv('../csv/sampled_data.csv')
```

```python
sampled_df.reset_index(inplace=True)
sampled_df.drop('index', axis=1, inplace=True)
sampled_df.head()
```

```python
def download_image(url, save_path):
    try:
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            with open(save_path, 'wb') as file:
                for chunk in response.iter_content(1024):
                    file.write(chunk)
            print(f"Downloaded: {save_path}")
        else:
            print(f"Failed to download {url}: Status code {response.status_code}")
    except Exception as e:
        print(f"Error downloading {url}: {e}")
```

```python
for index, row in sampled_df.iterrows():
    image_url = row['image_nutrition_url']
    image_filename = os.path.join('../image_data', f"image_{index}.jpg")
    download_image(image_url, image_filename)
```

```python
path = []
for i in range(0,1000):
  filepath = f'../image_data/image_{i}.jpg'
  path.append(filepath)
```

```python
sampled_df['path'] = path
```
